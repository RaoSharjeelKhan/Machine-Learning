```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import pandas as pd
import numpy as np
import re

url='https://www.zameen.com/Rentals/Rawalpindi-41-1.html'
url1='https://www.zameen.com/Rentals/Rawalpindi-41-2.html'
url2='https://www.zameen.com/Rentals/Rawalpindi-41-3.html'
url3='https://www.zameen.com/Rentals/Rawalpindi-41-4.html'
url4='https://www.zameen.com/Rentals/Rawalpindi-41-5.html'
url5='https://www.zameen.com/Rentals/Rawalpindi-41-6.html'
url6='https://www.zameen.com/Rentals/Rawalpindi-41-7.html'
page=requests.get(url)
page1=requests.get(url1)
page2=requests.get(url2)
page3=requests.get(url3)
page4=requests.get(url4)
page5=requests.get(url5)
page6=requests.get(url6)
```

```python
soup =BeautifulSoup(page.content,'html.parser')
soup1=BeautifulSoup(page1.content,'html.parser')
soup2=BeautifulSoup(page2.content,'html.parser')
soup3=BeautifulSoup(page3.content,'html.parser')
soup4=BeautifulSoup(page4.content,'html.parser')
soup5=BeautifulSoup(page5.content,'html.parser')
soup6 =BeautifulSoup(page6.content,'html.parser')
s=[soup,soup1,soup2,soup3,soup4,soup5,soup6]
b=[]
for st in s:
  lt=st.find_all('article',class_="f0349ab4")
  for li in lt:  
     price=li.find('span',class_='f343d9ce').text
     loc=li.find('div',class_='_162e6469').text     
     span=str(li.find_all('span',class_='_984949e5'))
     pattern=re.compile(r'>(\d)|(\d)[a-zA-Z]+<')
     matches=pattern.finditer(span)
     a=[]   
     for match in matches:
         a.append(match.group(1))   
     arr=[price,*a,loc]
     b.append(arr)    
     print(arr)
```

```python
df=pd.DataFrame(b,columns=['Price','Bedrooms','Bathrooms','Area','Location'])
print('Shape of the DataFrame: ',df.shape)
df.head()
```
